```
def newtons(x):
    print(np.cos(x))
    print(np.sin(x))
    x_next = x - (x - np.cos(x))/(1 + np.sin(x))
    return x_next
```

```
def iteration(x_0,n):
    x = x_0
    for i in range(0,n):
        x = newtons(x)
        print('Minimizer of the given function at %d th iteration is %.5f ' %(i,x))
    return x
```

```
x_0 = 0.8; #Initial value
n = 6; # Iterations
x_min = iteration(x_0,n)
print('Minimizer of the given function is %.5f ' %(x_min))

```
