```
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
#Import accuracy_score
from sklearn.metrics import accuracy_score
#Import numpy
import numpy as np
#Import pyplot from matplotlib
import matplotlib.pyplot as plt
#Import svm and datasets from sklearn
from sklearn import datasets
iris=datasets.load_iris()
```

```
X=iris.data[:, :2]
Y=iris.target
X0 = X[:, 0]
X1 = X[:, 1]
```

```
def make_meshgrid(x, y, h=.02):
#x_min :Minimum value in x axis; x_max :Maximum value in x axis
  x_min, x_max = x.min() - 1, x.max() + 1
#y_min :Minimum value in y axis; y_max :Maximum value in y axis
  y_min, y_max = y.min() - 1, y.max() + 1
  xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h)) 
  return xx, yy
```

```
xx, yy = make_meshgrid(X0, X1)
```

```
model = LinearDiscriminantAnalysis()
model.fit(X,Y)
```

```
pred = model.predict(np.c_[xx.ravel(), yy.ravel()])
#Find predicted labels
pred_values = model.predict(np.c_[X0.ravel(), X1.ravel()])
#Find accuracy score
accuracy_value=accuracy_score(Y, pred_values)
#Reshape the data
pred = pred.reshape(xx.shape)
```

```

plt.scatter(X0, X1, c=Y, cmap='prism', s=20,edgecolors='k')
#Set limit for x axis
plt.xlim(xx.min(), xx.max())
#Set limit for y axis
plt.ylim(yy.min(), yy.max())
#Set X label with accuracy score
x_label='Sepal length \n\n Accuracy = '+str("%.2f" % accuracy_value)
#Put label in x axis
plt.xlabel(x_label)plt.contourf(xx, yy, pred, cmap=plt.cm.coolwarm)
'''
Scatter the given labels in the graph
cmap : Colour map selection
s : Size of the points
edgecolors : Color of the points
refer : https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html
'''
#Put label in y axis
plt.ylabel("Sepal width")
#Set the graph title
plt.title('Linear Discriminant Analysis')
```
