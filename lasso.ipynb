```
#Import matplotlib
import matplotlib.pyplot as plt
#Import boston house prize dataset
from sklearn.datasets import load_boston
#Import SelectFromModel class from feature_selection module in sklearn package
from sklearn.feature_selection import SelectFromModel
#Import Lasso class from linear_model module
from sklearn.linear_model import Lasso
```

```
# Load the boston dataset.
boston = load_boston()
#Extract data and target from the dataset
X, y = boston['data'], boston['target']
#Get all feature names
all_feature_names=boston.feature_names
print("\ninitial features\n")
features_count=0
for feature in all_feature_names:
  features_count+=1
  print(str(features_count)+" "+feature)
#Initialize alpha value
alpha_value =1
```

```
#Plot the graphs untill the value of alpha exceeds 10
while (alpha_value<=10):
  clf = Lasso(alpha = alpha_value)
  sfm = SelectFromModel(clf)
  sfm.fit(X, y)
  n_features = sfm.transform(X).shape[1]
  objects = ('Before LASSO', 'After LASSO')
  count = [features_count,n_features]
  plt.bar(objects, count, align='center')
  plt.ylabel('Features count')
  #Set the title
  plt.title('Feature selection using LASSO (alpha = '+str(alpha_value)+' )')
  #Display the graph
  plt.show()
  print('The selected features when Alpha = '+str(alpha_value) +" are \n")
  #For getting the selected features after fitting
  #indices=True for getting the feature indices
  selected_feature_indices=sfm.get_support(indices=True)
  #Print the selected features
  count=1
  list=[]
  for index in selected_feature_indices:
    print(str(count)+" "+all_feature_names[index])
    list[i]=all_feature_names[index]
    count+=1
  print(*list)
  #Update the alpha_value
  alpha_value+=3
```
