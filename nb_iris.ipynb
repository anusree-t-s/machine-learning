```
import numpy as np
import sklearn.metrics
from  sklearn import  datasets
iris=datasets.load_iris()
```

```
x=iris.data
y=iris.target
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.5)
```

# **gaussian**

```
from sklearn.naive_bayes import GaussianNB
model1=GaussianNB()
model1.fit(x_train,y_train)
y_predict1=model1.predict(x_test)
print(sklearn.metrics.f1_score(y_test,y_predict1,average='micro'))
```

# **multinomial**

```
from sklearn.naive_bayes import MultinomialNB
model2 = MultinomialNB()
#model2=MultinomialNB(alpha=1.0, class_prior=None, fit_prior=True)
model2.fit(x_train,y_train)
y_predict2=model2.predict(x_test)
```

```
print(sklearn.metrics.f1_score(y_test,y_predict2,average='micro'))
```

# **compliment**

```
from sklearn.naive_bayes import ComplementNB
model3= ComplementNB()
#model3=ComplementNB(alpha=1.0, class_prior=None, fit_prior=True, norm=False)
model3.fit(x_train,y_train)
y_predict3=model3.predict(x_test)
print(sklearn.metrics.f1_score(y_test,y_predict3,average='micro'))
```

# **bernouli**

```
from sklearn.naive_bayes import BernoulliNB
model4= BernoulliNB()
model4.fit(x_train,y_train)
y_predict4=model4.predict(x_test)
print(sklearn.metrics.f1_score(y_test,y_predict4,average='micro'))
```
