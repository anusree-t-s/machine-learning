```
#Import make_circles class from sklearn
from sklearn.datasets import make_circles
#Import SpectralClustering
from sklearn.cluster import SpectralClustering
#Import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```
X, y = make_circles(n_samples=3000, noise=0.002)
# scatter plot, dots colored by class value
plt.scatter(X[:,0], X[:,1])
#Plot x label
plt.xlabel('X axis values')
#Plot y label
plt.ylabel('Y axis values')
#Show the plotting
plt.show()
```

```
fig, sub = plt.subplots(3, 3,figsize=(10,9))
plt.subplots_adjust(wspace=0.7, hspace=0.6)
#Initialize the gamma value
gamma_value=100
for graph in sub.flatten():
  clustering =SpectralClustering(n_clusters=2,eigen_solver='arpack',affinity='rbf',gamma=gamma_value).fit(X)
#Plot x label
  graph.set_xlabel('X axis values')
#Plot y label
  graph.set_ylabel('Y axis values')
#Display the title
  title='Gamma = '+str("%.2f" % gamma_value)
  graph.set_title(title)
#Display the graph
  graph.scatter(X[:,0],X[:,1], c=clustering.labels_, cmap='prism')
#Update the gamma value
  gamma_value+=20
```
