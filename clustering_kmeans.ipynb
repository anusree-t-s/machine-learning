```
from sklearn.datasets.samples_generator import make_blobs
#Import pyplot from matplotlib
import matplotlib.pyplot as plt
#Import KMeans
from sklearn.cluster import KMeans
X, y = make_blobs(n_samples=50, centers=2, n_features=2)
# scatter plot, dots colored by class value
plt.scatter(X[:,0], X[:,1])
#Plot x label
plt.xlabel('X axis values')
#Plot y label
plt.ylabel('Y axis values')
#Show the plotting
plt.show()
#Initialize the k value
fig, sub = plt.subplots(3, 3,figsize=(10,9))
plt.subplots_adjust(wspace=0.3, hspace=0.35)
kvalue=5
for graph in sub.flatten():
  kmeans = KMeans(n_clusters=kvalue)
#Train the object with the data
  kmeans.fit(X)
#Plot x label
  graph.set_xlabel('X axis values')
#Plot y label
  graph.set_ylabel('Y axis values')
  title='K = '+str(kvalue)
  graph.set_title(title)
#Display the graph
  graph.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='prism')
#update k value
kvalue+=1
```

```
c1,c2=np.random.randn(2,1),np.random.randn(2,1)
```

```
print(c1)
```

```
plt.scatter(X[0,:],X[1,:],c=['r']*100+['g']*100)
plt.scatter(c1[0,0],c1[1,0],c='b')
plt.scatter(c2[0,0],c2[1,0],c='cyan')
plt.show()
```

```
dif1=X-c1
d1=(dif1*dif1).sum(axis=0)
```

```

dif2=X-c2
d2=(dif2*dif2).sum(axis=0)
```

```
cluster1,cluster2=[],[]
for i ,(dist1,dist2) in enumerate(zip(d1,d2)):
  if dist1<dist2:
    cluster1.append(X[:,i])
  else:
    cluster2.append(X[:,i])
cluster1=np.array(cluster1).T
cluster2=np.array(cluster2).T
```

```
print(cluster2.shape)
```

```
c1=cluster1.mean(axis=1)
c1=c1.reshape(2,1)
c2=cluster2.mean(axis=1)
c2=c2.reshape(2,1)
```

```
print(c2.shape)
```
